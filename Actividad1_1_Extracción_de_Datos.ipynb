```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Filtro por objeto
filtro1=df[df['NOMBRE_VENDEDOR']=='LETICIA RAMIREZ HERNANDEZ']
filtro1
```

```python
#DESCARGA
```

```python
#filro por filas
#filtro2=df.iloc[3:10,:] # filas consecutivas
filtro2=df.iloc[[3,8,100],:] #filas no consecutivas
filtro2
```

```python
#Filtro por columnas
#filtro3=df.iloc[:,1:3] #columas consecutivas
filtro3=df.iloc[:, [4,6,10]] #columnas no consecutivas
filtro3
```

```python
#Paso2: Aplicamos filtro de fials vs columnas
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col=2)
df1
filtro4=df1.loc[['DIEGO ARMANDO LEON DURAN'],['NOMBRE_CLIENTE']] #se filtra nombre del cliente
filtro4
```

```python
#filtro por cabecera
filtro5=df.head()
filtro5
```

```python
#filtro por cola
filtro6=df.tail()
filtro6
```

```python
filtro7=df[df['PRECIO_UNITARIO']>400] #se filtra el precio unitario mayor a 400
filtro7
```

```python
#Filtros conectores lógicps
#Filtro y
filtro8=df[(df['PRECIO_UNITARIO']>400) & (df['NOMBRE_VENDEDOR']=='LETICIA RAMIREZ HERNANDEZ')] # se filtra precio unitario mayor a 400 y el nombre del venedor de LETICIA RAMIREZ HERNANDEZ
filtro8
```

```python
#Filtro or
filtro9=df[(df['PRECIO_UNITARIO']>1000) | (df['NOMBRE_VENDEDOR']=='DIEGO ARMANDO LEON DURAN')] #se filtra precio unitario mayor a 1000 y el nombre del venedor de DIEGO ARMANDO lEON DURAN
filtro9
```

```python
#Filtro not
filtro10=df[~(df['PRECIO_UNITARIO']>1000) &~ (df['NOMBRE_CLIENTE']=='OMAR RODRIGUEZ GOMEZ')]
filtro10
```
